```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
import seaborn as sns
```

Задание №1. Проведите корреляционный анализ данных из файла 'ex1.csv' по Пирсону

```python
# считывание файла
df1 = pd.read_csv('ex1.csv')
df1
```

```python
# КРИТЕРИЙ 1

# проверка, что все элементы - численного типа

# построение гистограмм распределения

# построение "ящика с усами" для изучения выбросов

# построение диаграммы рассеяния

# вывод о целесообразности использования корреляции Пирсона с обоснованием на основе предшествующих графиков
```

```python
if df1.dtypes['x'].kind in 'biufc' and df1.dtypes['y'].kind in 'biufc':
    print("Все элементы имеют численный тип")
else:
    print("Некоторые элементы не являются числовыми")

plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
plt.hist(df1['x'], bins=10, color='skyblue', edgecolor='black')
plt.title('Гистограмма распределения x')
plt.xlabel('x')
plt.ylabel('Частота')

plt.subplot(1, 2, 2)
plt.hist(df1['y'], bins=10, color='lightgreen', edgecolor='black')
plt.title('Гистограмма распределения y')
plt.xlabel('y')
plt.ylabel('Частота')

plt.tight_layout()
plt.show()

plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
sns.boxplot(x=df1['x'])
plt.title('Ящик с усами для x')

plt.subplot(1, 2, 2)
sns.boxplot(x=df1['y'])
plt.title('Ящик с усами для y')

plt.tight_layout()
plt.show()

plt.figure(figsize=(8, 6))
plt.scatter(df1['x'], df1['y'], color='purple')
plt.title('Диаграмма рассеяния между x и y')
plt.xlabel('x')
plt.ylabel('y')
plt.grid()
plt.show()

correlation = df1.corr(method='pearson')

if abs(correlation.loc['x', 'y']) > 0.5:
    print("Целесообразно использовать корреляцию Пирсона, так как графики показывают наличие линейной зависимости между x и y")
else:
    print("Корреляция Пирсона может быть нецелесообразна из-за слабой линейной зависимости между x и y")
correlation

#Наблюдений много, и данные распределены нормально, можем применять кореляцию Пирсона
```

```python
# КРИТЕРИЙ 2

# расчет коэффициента Пирсона

# расчет t-критерия

# вывод о степени корреляции величин и статистической значимости их взаимосвязи 
```

```python
correlation_coefficient, p_value = stats.pearsonr(df1['x'], df1['y'])
print(f"Коэффициент корреляции Пирсона: {correlation_coefficient}")
print(f"p-значение: {p_value}")

t_statistic = correlation_coefficient * np.sqrt((len(df1['x']) - 2) / (1 - correlation_coefficient**2))
print(f"t-статистика: {t_statistic}")

alpha = 0.05
if p_value < alpha:
    print("Отвергаем нулевую гипотезу: корреляция статистически значима")
else:
    print("Принимаем нулевую гипотезу: корреляция не статистически значима")

if abs(correlation_coefficient) < 0.3:
    correlation_strength = "Слабая"
elif abs(correlation_coefficient) < 0.7:
    correlation_strength = "Умеренная"
else:
    correlation_strength = "Сильная"

print(f"Степень корреляции: {correlation_strength} ({correlation_coefficient})")
```

Задание №3. Проведите линейный регрессионный анализ данных из файла 'ex3.csv'

```python
# считывание файла
df13 = pd.read_csv('ex3.csv')
df13
```

```python
#Проверим нормальность данных
def check_normality(data):
    stat_x, p_value_x = stats.shapiro(data['x'])
    stat_y, p_value_y = stats.shapiro(data['y'])

    alpha = 0.05

    print(f'Тест Шапиро-Уилса для x: Statistics={stat_x}, p-value={p_value_x}')
    if p_value_x < alpha:
        print("Отвергаем нулевую гипотезу: данные в наборе x не распределены нормально")
    else:
        print("Принимаем нулевую гипотезу: данные в наборе x распределены нормально")

    print(f'Тест Шапиро-Уилса для y: Statistics={stat_y}, p-value={p_value_y}')
    if p_value_y < alpha:
        print("Отвергаем нулевую гипотезу: данные в наборе y не распределены нормально")
    else:
        print("Принимаем нулевую гипотезу: данные в наборе y распределены нормально")

    plt.figure(figsize=(12, 6))

    plt.subplot(1, 2, 1)
    stats.probplot(data['x'], dist="norm", plot=plt)
    plt.title('Q-Q plot для x')

    plt.subplot(1, 2, 2)
    stats.probplot(data['y'], dist="norm", plot=plt)
    plt.title('Q-Q plot для y')

    plt.show()

check_normality(df13)
```

```python
# КРИТЕРИЙ 5

# построение диаграммы рассеяния

# вывод о целесообразности использования корреляции Пирсона

# расчет коэффициента Пирсона
```

```python
plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
plt.hist(df13['x'], bins=10, color='skyblue', edgecolor='black')
plt.title('Гистограмма распределения x')
plt.xlabel('x')
plt.ylabel('Частота')

plt.subplot(1, 2, 2)
plt.hist(df13['y'], bins=10, color='lightgreen', edgecolor='black')
plt.title('Гистограмма распределения y')
plt.xlabel('y')
plt.ylabel('Частота')

plt.tight_layout()
plt.show()

plt.figure(figsize=(8, 6))
plt.scatter(df13['x'], df13['y'], color='blue')
plt.title('Диаграмма рассеяния между x и y')
plt.xlabel('x')
plt.ylabel('y')
plt.grid()
plt.show()

pearson_coefficient, p_value = stats.pearsonr(df13['x'], df13['y'])
print(f"Коэффициент корреляции Пирсона: {pearson_coefficient}")
print(f"p-значение: {p_value}")

alpha = 0.05
if p_value < alpha:
    print("Отвергаем нулевую гипотезу: корреляция статистически значима")
else:
    print("Принимаем нулевую гипотезу: корреляция не статистически значима")

if abs(pearson_coefficient) < 0.3:
    correlation_strength = "Слабая"
elif abs(pearson_coefficient) < 0.7:
    correlation_strength = "Умеренная"
else:
    correlation_strength = "Сильная"

print(f"Степень корреляции: {correlation_strength} ({pearson_coefficient})")
print("Корреляция Пирсона подходит для анализа линейной зависимости между переменными")
```

```python
# КРИТЕРИЙ 6

# расчет коэффициентов линейной регрессии

# построение диаграммы рассеяния прогнозных и реальных значений на одном графике
```

```python
import statsmodels.api as sm
X = sm.add_constant(df13['x'])
model = sm.OLS(df13['y'], X).fit()
print(f'Коэффициенты модели - const: {model.params.iloc[0]}, x: {model.params.iloc[1]}')
print(f'Уравнение линейной регрессии: y = {model.params.iloc[0]}  {model.params.iloc[1]}*x')

plt.figure(figsize=(8, 6))
plt.scatter(df13['x'], df13['y'], color='blue', label='Реальные значения')
plt.plot(df13['x'], model.predict(X), color='red', label='Прогнозные значения', linewidth=2)
plt.title('Сравнение реальных и прогнозных значений')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.grid()
plt.show()
```

```python
# КРИТЕРИЙ 11

# построение диаграмм рассеяния (x1/y, x2/y, x1/x2)

# вывод о корректности использования модели множественной регрессии

# расчет коэффициентов корреляции

# вывод о необходимости устранения какого-либо фактора
```

```python
plt.figure(figsize=(10, 6))
plt.subplot(1, 3, 1)
plt.scatter(df14['x1'], df14['y'], color='blue')
plt.title('Диаграмма рассеяния: x1 и y')
plt.xlabel('x1')
plt.ylabel('y')

plt.subplot(1, 3, 2)
plt.scatter(df14['x2'], df14['y'], color='green')
plt.title('Диаграмма рассеяния: x2 и y')
plt.xlabel('x2')
plt.ylabel('y')

plt.subplot(1, 3, 3)
plt.scatter(df14['x1'], df14['x2'], color='red')
plt.title('Диаграмма рассеяния: x1 и x2')
plt.xlabel('x1')
plt.ylabel('x2')


plt.tight_layout()
plt.show()

X = df14[['x1', 'x2']]
X = sm.add_constant(X)
y = df14['y']

model = sm.OLS(y, X).fit()

model_summary = model.summary()
print(model_summary)

#Модель множественной регрессии корректно применима, поскольку отсутствует мультиколлинеарность
#R-squared =0.996 указывает на то, что модель объясняет 99.6% вариации зависимой переменной y
#Это свидетельствует о высокой точности модели
#Оба фактора (x1 и x2) статистически значимы и вносят вклад в модель
# Мультиколлинеарности между факторами нет (корреляция между x1 и x2 низкая)
#Нет необходимости исключать какой-либо фактор
```

```python
# КРИТЕРИЙ 12

# расчет стандартных коэффицентов регрессии
```

```python
X = df14[['x1', 'x2']]
y = df14['y']

X_std = (X - X.mean()) / X.std()
y_std = (y - y.mean()) / y.std()

model_standardized = sm.OLS(y_std, X_std).fit()
standardized_coefficients = model_standardized.params
print(f"Коэффициенты стандартной регрессии - '{model_standardized.params.index[0]}' : {model_standardized.params.iloc[0]}, '{model_standardized.params.index[1]}' : {model_standardized.params.iloc[1]}")
```

```python
# КРИТЕРИЙ 13

# расчет коэффициентов "чистой" регрессии

# построение диаграмм рассеяния прогнозных и реальных значений на одном графике
```

```python
X = df14[['x1', 'x2']]
y = df14['y']
X = sm.add_constant(X)
model = sm.OLS(y, X).fit()
print(f"Коэффициенты чистой регрессии - '{model.params.index[0]}' : {model.params.iloc[0]}, '{model.params.index[1]}' : {model.params.iloc[1]}, '{model.params.index[2]}' : {model.params.iloc[2]}")

model = sm.OLS(y, X).fit()

y_pred = model.predict(X)

plt.figure(figsize=(8, 6))
plt.scatter(y, y_pred)
plt.plot([y.min(), y.max()], [y.min(), y.max()], 'r--', lw=2)
plt.title('Прогнозные vs Реальные значения')
plt.xlabel('Реальные значения (y)')
plt.ylabel('Прогнозные значения (y_pred)')
plt.grid()
plt.show()
```

```python
# КРИТЕРИЙ 14

# расчет индекса множественной корреляции

# расчет F-критерия

# вывод о статистической значимости уравнения в целом
```

```python
r_sqr = model.rsquared
corr_index = r_sqr ** 0.5

print("Коэффициент детерминации (R^2):", r_sqr)
print("Индекс множественной корреляции (R):", corr_index)
f_statistic = model.fvalue
print(f"F-критерий: {f_statistic}")
#На основе высокого значения F-критерия и высокого коэффициента детерминации  можно сделать вывод, что уравнение регрессии статистически значимо
```
